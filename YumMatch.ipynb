# Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://jetson-datathon-sample-data.s3.us-east-2.amazonaws.com/jetson-sample-data.csv')
df.date = pd.to_datetime(df.date)
df
```

```python
result = df.groupby(['order_id','date'])['item_name'].apply(','.join).reset_index()
```

```python
result['item_name'] = result['item_name'].apply(lambda x: x.split(','))
```

```python
result
```

# Popular Combos

```python
import itertools

combinations_list = []
for row in result.item_name:
    combinations = list(itertools.combinations(row, 2))
    combinations_list.append(combinations)
```

```python
combination_counts = pd.Series(combinations_list).explode().reset_index(drop=True)
combination_counts = combination_counts.value_counts().reset_index()
combination_counts.columns = ["Sets", "Count"]
combination_counts
```

```python
popular = combination_counts.head(50)
popular
```

```python
import pickle
pickle.dump(popular, open('/content/popular','wb'))
```

# Combo Recommendation

```python
df_temp = df.copy().drop(["client_id", "price", "date", "quantity"], axis=1)
onehot = df_temp.pivot_table(index='order_id', columns='item_name', aggfunc=len, fill_value=0)
onehot = onehot>0
onehot
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# compute frequent items using the Apriori algorithm
frequent_itemsets = apriori(onehot, min_support = 0.001, max_len = 2, use_colnames=True)

# compute all association rules for frequent_itemsets
rules = association_rules(frequent_itemsets)
rules.sort_values('confidence', ascending=False).reset_index(drop=True)
```

```python
rules.columns
```

```python
recommend = rules[['antecedents', 'consequents', 'confidence']].copy()
```

```python
print(recommend)
```

```python
# Apply the function to the 'fset' column and convert to strings
recommend['antecedents'] = recommend['antecedents'].map(str)
recommend['consequents'] = recommend['consequents'].map(str)
```

```python
# define the conversion function
def convert_str(val):
    if isinstance(val, str) and val.count("'") == 2:
        return val.replace("'", '"')
    else:
        return val

# apply the conversion function to all elements of the dataframe
recommend = recommend.applymap(convert_str)
```

```python
recommend
```

```python
import re
pattern = r'frozenset\(\{?"(.*?)"?\}?\)'
recommend['antecedents'] = recommend['antecedents'].str.extract(pattern)
recommend['consequents'] = recommend['consequents'].str.extract(pattern)
```

```python
recommend
```

```python
pickle.dump(recommend.to_dict(), open('/content/recommend_dict','wb'))
```
